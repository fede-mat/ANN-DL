```python
import numpy as np

# Load the .npz file
data = np.load('training_set.npz')
print(data.files)
for key in data.files:
    print(f"{key}: shape={data[key].shape}, dtype={data[key].dtype}")
    
images = data['images']     # Access the images array
labels = data['labels']     # Access the labels array
import matplotlib.pyplot as plt
```

```python


# Display a sample image (assuming the images are in the format (num_images, height, width))
fig, axs = plt.subplots(4,7)
for y in range(7):
	for x in range(4):
		index = np.random.randint(0,13759)
		while (labels[index] != y) | (np.array_equal(images[index],images[12760]) == True) | (np.array_equal(images[index],images[13580]) == True):
			index = np.random.randint(0,13759)
		axs[x,y].imshow(images[index], cmap='gray')  
# plt.hist(labels)
```
